```python
def sum_multiples_3_or_5(n):
    sum = 0
    for i in range(n):
        if i % 3 == 0 or i % 5 == 0:
            sum += i
    return sum

# Example usage:
N = 1000
result = sum_multiples_3_or_5(N)
print(f"The sum of all multiples of 3 or 5 below {N} is: {result}")
```

```python
def sum_even_fibonacci(n):
    a, b = 1, 1
    sum_even = 0
    while a <= n:
        if a % 2 == 0:
            sum_even += a
        a, b = b, a + b
    return sum_even

# Example usage:
N = 4000000
result = sum_even_fibonacci(N)
print(f"The sum of even-valued Fibonacci terms not exceeding {N} is: {result}")
```

```python
def largest_prime_factor(n):
    # Initialize the largest prime factor variable
    largest_prime = 0

    # Check divisibility by 2
    while n % 2 == 0:
        largest_prime = 2
        n //= 2  # Divide n by 2 repeatedly until it is odd

    # Check divisibility by odd numbers from 3 onwards
    factor = 3
    while factor * factor <= n:
        while n % factor == 0:
            largest_prime = factor
            n //= factor
        factor += 2  # Increment to the next odd number

    # If n is a prime number greater than 2
    if n > 2:
        largest_prime = n

    return largest_prime

# Example usage:
N = 600851475143
result = largest_prime_factor(N)
print(f"The largest prime factor of {N} is: {result}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Binomial distribution parameters
n = 100  # number of trials
p = 0.5  # probability of success

# Normal distribution parameters
mu = 0    # mean
sigma = 1 # standard deviation

# Poisson distribution parameter
lambda_ = 5  # rate parameter

# Generate random numbers from each distribution
data_binomial = np.random.binomial(n, p, 1000)
data_normal = np.random.normal(mu, sigma, 1000)
data_poisson = np.random.poisson(lambda_, 1000)

# Plot histograms
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.hist(data_binomial, bins=30, density=True, alpha=0.7, color='b')
plt.title('Binomial Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.hist(data_normal, bins=30, density=True, alpha=0.7, color='g')
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.hist(data_poisson, bins=30, density=True, alpha=0.7, color='r')
plt.title('Poisson Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)

plt.tight_layout()
plt.show()
```
